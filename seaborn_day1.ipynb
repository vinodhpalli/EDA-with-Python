## Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

## univariate plot

```python
##univariate plot-->we need to take only one axis as input(x or y)
```

```python
# histogram
sns.histplot(data=tips,x='total_bill',bins=20,kde=True)
# kde-->kernel density estimation
plt.title("Histogram of total bill")
plt.show()
```

```python
sns.histplot(data=tips,x='total_bill',bins=20,color='skyblue',edgecolor='red',kde=True)
plt.xlim(0,60)
plt.ylim(0,50)
plt.title("Histogram of total bill",fontsize=16)
plt.xlabel("Total Bill",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

## KDE Plot

 insted of drawing bars like Histogram,
KDE plot draws a smooth curve that represents how data points are distrubuted 

```python
sns.kdeplot(data=tips,x="total_bill",fill=True,color="skyblue",linewidth=2,linestyle='--')
plt.xlim(0,60)
plt.ylim(0,0.06)
plt.title("KDE plot of Total bill" ,fontsize=20)
plt.xlabel("Total Bill",fontsize=12)
plt.ylabel("Density",fontsize=12)
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

## Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot of total bill by day")
plt.show()
```
