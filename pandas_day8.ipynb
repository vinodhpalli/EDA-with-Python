```python
data={
    "Name":["A","B","C","D","E","F"],
    "Age":[25,30,35,40,45,50]
}
df=pd.DataFrame(data,index=[0,1,1,2,2,2])
df
```

## reset index

```python
df_reset=df.reset_index(drop=True)
df_reset
```

## keep="first"

```python
df_unique=df[df.index.duplicated()]
df_unique
```

```python

# drop duplicate index rows (keep first occurence)
df_unique=df[~df.index.duplicated(keep="first")]
df_unique
```

# keep="last"Â¶

```python
# drop duplicate index rows (keep last occurence)
df_unique=df[~df.index.duplicated(keep="last")]
df_unique
```

# keep="False"

```python
# drop all rows with duplicates 
df_unique=df[~df.index.duplicated(keep=False)]
df_unique
```

## Group by(),Agg()

```python
data={
    "Empid":[1,2,3,4,5,6],
    "Name":["A","B","C","D","E","F"],
    "Department":["Sales","HR","Sales","IT","HR","IT"],
    "Salary":[50000,60000,50000,70000,60000,70000]
}
df=pd.DataFrame(data)
df
```

```python
summary=df.groupby("Department").agg(
    Avg_Salary=("Salary","mean"),
    Total_salary =("Salary","sum"),
    Employee_count=("Empid","count")
)
summary
     
```

```python
# find maximum salary per department
max_salary=df.groupby("Department")["Salary"].max()
max_salary
```

```python
# find minimum salary per department
mix_salary=df.groupby("Department")["Salary"].min()
max_salary
```

```python
multi=df.groupby("Department")["Salary"].agg(
    ['mean','sum','min','max','count','median']
)
multi
```

```python
a=df.groupby(["Department","Salary"]).size()
a
```
