## üîπ PART 4: INDEX OPERATIONS


```python
import pandas as pd
```

```python
data = {
    "Employee_ID": ["E1", "E2", "E3", "E4"],
    "Name": ["Asha", "Ravi", "Neha", "Arjun"],
    "Department": ["IT", "HR", "IT", "Finance"],
    "Gender": ["F", "M", "F", "M"],
    "Salary": [60000, 50000, 75000, 65000],
    "Score": [78, 85, 92, 88]
}

df = pd.DataFrame(data)
df
```

```python
#üîü Change the Index
#Input

 #Column name = Employee_ID
# Check uniqueness first
#print(df["Employee_ID"].is_unique)

# Then set index
print(df["Employee_ID"].is_unique)
```

```python
#1Ô∏è‚É£1Ô∏è‚É£ Restore Default Index
#Input
 #DataFrame with Employee_ID as index
#Expected Output
# Default numeric index (0,1,2,3)
#Task
 #Reset the index.

df = df.reset_index(drop=True)
df
```

```python
#1Ô∏è‚É£2Ô∏è‚É£ Index Sorting
#Input
 #Index order = E3, E1, E4, E2
#Expected Output
#E1, E2, E3, E4
#Task
 #Sort the DataFrame by index in ascending order.
df = df.sort_index()
df
```

## üîπ PART 5: SORTING ROWS


```python
#1Ô∏è‚É£3Ô∏è‚É£ Sort by Salary
#Input
 #Salary column
#Expected Output
 #Salaries arranged from highest to lowest
#Task
 #Sort employees by salary in descending order.
df = df.sort_values(by="Salary", ascending=False)
df
```

```python
#1Ô∏è‚É£4Ô∏è‚É£ Sort by Department and Salary
#Input
 #Department (A‚ÄìZ), Salary (High ‚Üí Low)
#Expected Output
 #Employees grouped by department and ordered by salary within each department
#Task
 #Perform multi-column sorting.
df = df.sort_values(by=["Department", "Salary"],
                    ascending=[True, False])
df
```

## üîπ PART 6: GROUPING DATA


```python
#1Ô∏è‚É£5Ô∏è‚É£ Department-wise Salary Total
#Input
 #Department and Salary columns
#Expected Output
#Department
#Total Salary
dept_salary_total = df.groupby("Department")["Salary"].sum()
print(dept_salary_total)
```

```python
#1Ô∏è‚É£6Ô∏è‚É£ Department and Gender Grouping
#Input
 #Department, Gender, Salary
#Expected Output
 #Grouped salary summary
#Task
 ##Group employees by department and gender
grouped = df.groupby(["Department", "Gender"])["Salary"].sum()
print(grouped)
```

```python
#1Ô∏è‚É£7Ô∏è‚É£ Multiple Aggregations
#Input
 #Department and Salary
#Expected Output
#Department
#Minimum Salary
#Maximum Salary
#Task
# Find minimum and maximum salary per department.
result = df.groupby("Department")["Salary"].agg(["min", "max"])
print(result)
```

## üîπ PART 7: SUMMARY TABLES


```python
#1Ô∏è‚É£8Ô∏è‚É£ Pivot Table
#Input
 #Department and Salary
#Expected Output
 #Average salary per department
##Create a pivot table.
pivot = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="mean"
)

print(pivot)
```

```python
#1Ô∏è‚É£9Ô∏è‚É£ Cross Tabulation
#Input
 #Gender and Department
#Expected Output
 #Count of employees for each gender in each department
#Task
 #Create a cross-tabulation table.
crosstab = pd.crosstab(df["Gender"], df["Department"])
print(crosstab)
```

## üîπ PART 8: FINAL COMBINED TASK


```python
#2Ô∏è‚É£0Ô∏è‚É£ Complete Data Processing
#Input
 #The DataFrame df
#Expected Output
 #Sorted, department-wise salary summary table
#Task
 #Apply filtering, grouping, aggregation, sorting, and index reset.
result = (
    df
    .dropna(subset=["Salary"])
    .assign(Salary=lambda x: pd.to_numeric(x["Salary"]))
    .groupby("Department", as_index=False)
    .agg(
        Total_Salary=("Salary", "sum"),
        Average_Salary=("Salary", "mean"),
        Employee_Count=("Salary", "count")
    )
    .sort_values(by="Total_Salary", ascending=False)
    .reset_index(drop=True)
)

print(result)

```
