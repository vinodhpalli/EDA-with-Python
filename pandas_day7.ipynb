# Droping Duplicates in Row Index

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data={
    "Name":["A","B","B","D","D","F"],
    "Age":[25,30,30,40,45,50],
    "City":["New york","Paris","Paris","Berlin","Paris","London"],
    "Salary":[50000,60000,70000,80000,90000,100000]
}
df=pd.DataFrame(data)
df
```

```python
df_duplicates=df.duplicated(keep="first")
df_duplicates
```

```python
df_duplicates=df.duplicated(keep="last")
df_duplicates
```

```python
df_duplicates=df.duplicated(keep=False)
df_duplicates
```
