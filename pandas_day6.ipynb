# boolean row selection by values

```python
import pandas as pd
```

```python
data={
    "Name":["A","B","C","D","E","F"],
    "Age":[25,30,35,40,45,50],
    "City":["New york","Paris","London","Berlin","Paris","London"],
    "Salary":[50000,60000,70000,80000,90000,100000]
}
df=pd.DataFrame(data)
df
```

```python
df[df["Age"]>35]
```

```python
df[df["Age"]<35]
```

```python
df[df["City"]=="Paris"]
```

```python
df[df["Age"]==35]
```

# And conditon(&)

```python

df[(df["Age"]>35) & (df["City"]=="Paris")]
```

# or conditon(|)

```python
# or conditon(|)
df[(df["Age"]<35) | (df["City"]=="Paris")]
```

# is in () condition

```python
df[df["City"].isin(["Paris","Hyderabad"])]
                    
```

# not in (~)---> (with is in())

```python
df[~df["City"].isin(["Paris","Hyderabad"])] # give other than we mentioned citys
```

# str.contains

```python
df[df["City"].str.contains("on")]
```

```python
#  string not contains
df[~df["City"].str.contains("on")]
```

# Between

```python
df[df["Age"].between(30,45)]
```

#  notBetween

```python
df[~df["Age"].between(30,45)]
```

# Startswith


```python
df[df["City"].str.startswith("N")]
```

#  N0t StartswithÂ¶

```python
df[~df["City"].str.startswith("N")]
```

# Ends with

```python
df[df["City"].str.endswith("n")]
```

# Not Ends with

```python
df[~df["City"].str.endswith("n")]
```

```python
# age>30,city=london,salary>70000
df[(df["Age"]>30) & (df["City"]=="London") & (df["Salary"]>70000)]
```
