## logical operator

```python
import numpy as np
```

```python
a=np.array([10,20,30,40])
print(np.logical_and(a>15,a<35))
print(np.logical_or(a<15,a>35))
print(np.logical_not(a>25))
```

## sin,cos

```python
angles_deg=np.array([0,20,45,90])

#convert degrees toradians
angles_rad=np.deg2rad(angles_deg)
print("degrees:",angles_deg)
print("radians:",angles_rad)

#sin value
sin_val=np.sin(angles_rad)
print("sin value:",sin_val)
```

```python
angles_deg=np.array([[0,20],
                    [45,90]])
angles_rad=np.deg2rad(angles_deg)
print("degrees:",angles_deg)
print("radians:",angles_rad)

#sin value
sin_val=np.sin(angles_rad)
print("sin value:",sin_val)

#cos value
cos_val=np.cos(angles_rad)
print("cos value:",cos_val)
```

```python
angles=np.array([0,np.pi/4,np.pi/3,np.pi/2])
cos_val=np.cos(angles)
print("degrees:",angles)
print("cos value:",cos_val)
```

# numpy statistics

## sum()

```python
arr=np.array([1,2,3,4,5])
total=np.sum(arr)
print(total)
```

```python
a=np.array([[1,2,3],
            [4,5,6]])
total=np.sum(a)
print(total)

#sum of rows
print("sum of rows:",np.sum(a,axis = 1))
print("sum of coloumns:",np.sum(a,axis=0))
```

## min()

```python
arr=np.array([1,2,3,4,5])
minimum=np.min(arr)
print(minimum)
```

```python
a=np.array([[1,2,3],
            [4,5,6]])
minimum=np.min(a)
print(minimum)
print("minimum of rows:",np.min(a,axis = 1))
print("minimum of coloumns:",np.min(a,axis=0))
```

## max()

```python
a=np.array([[1,2,3],
            [4,5,6]])
maximum=np.max(a)
print(maximum)
print("maximum of rows:",np.min(a,axis = 1))
print("maximum of coloumns:",np.min(a,axis=0))
```
