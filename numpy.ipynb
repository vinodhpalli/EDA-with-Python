```python
#1d array(single dimension)--->1 index
#vector-->single dimension array
a=np.array([1,2,3,4,5])
print(a)
print("dimension:",a.ndim)
print("shape:",a.shape)
print("size:",a.size)# it will give no of elements in a array
```

```python
##2d array(matrix)--->2 indeces(row,coloum)
#matrix---->2 dimensional array
b=np.array([[1,2,3],
           [4,5,6]])
print(b)
print("dimension:",b.ndim)
print("shape:",b.shape)
print("size:",b.size)#it will give no of elements in a array
```

```python
#3d array(cube or tensor)--->3 indeces(block/depth,row,coloum)
#cube or tensor--->3dimensional array
c=np.array([[[1,2],
            [3,4]],
           
           [[5,6],
            [7,8]]])
print(c)
print("dimension:",c.ndim)
print("shape:",c.shape)# it give separately 2 rows, 2 block ,2 columns
print("size:",c.size)#it will give no of elements in a array
```

### numpy builtin functions

```python
#np.zeros()-->create a array filled with 0s
#np.ones()--->>create a array filled with 1s
#np.arrange-->create an array with a range of numbers
#np.linspace-->create an array with a range of numbers
#np.full-->create an array with a specific value
#np.eye-->create an array with squared identity matrix
#np.diag-->create an diagnol array  from a list or extract diagnol
```

```python
print("zeros:\n",np.zeros((2,3)))
print("ones:\n",np.ones((2,3)))
print("arange:\n",np.arange(0,11))
print("linspace:\n",np.linspace(0,1,5))# it will generate 0 to 1 ,it will  generate 5 numbers
print("full:\n",np.full((2,2),9))
print("eye:\n",np.eye(3))
print("diag:\n",np.diag([1,2,3]))
```

## Numpy data types

```python
arr=np.array([-1,-2,-3],dtype=np.int32)
print(arr)
print(arr.dtype)
```

```python
arr1=np.array([1,2,3],dtype=np.uint32)
print(arr1)
print(arr1.dtype)
```

### float

```python
arr=np.array([1.5,2.5,3.5],dtype=np.float16)
print(arr)
```

### complex

```python
arr1=np.array([1+2j,2+3j,3+4j],dtype=np.complex64)
print(arr1)
print(arr1.dtype)
```

#### boolean(bool)

```python
#stores true or false
arr=np.array([True,False,True],dtype=bool)
print(arr)
print(arr.dtype)
```

### string/unicode

```python
#stores text values
#dtype="U1"----->unicode string per element
arr=np.array(["a","b","c"],dtype="U1")
print(arr)
```
