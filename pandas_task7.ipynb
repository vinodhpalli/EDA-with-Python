```python
import pandas as pd
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("uncleaned_customers (1).csv")
df
```

```python
#Access the Name column as a Series.

print("Columns in dataset:")
print(df.columns.tolist())


if "Name" in df.columns:
    name_series = df["Name"]
    print("\nType of extracted object:")
    print(type(name_series))
    
    print("\nData type of the Series:")
    print(name_series.dtype)
    
    print("\nFirst 5 values:")
    print(name_series.head())
else:
    print("\nColumn 'Name' not found.")
```

```python
#Select multiple columns: Name and PurchaseAmount.
selected_columns = df[["Name", "PurchaseAmount"]]

print("Type of selected object:")
print(type(selected_columns))

print("\nFirst 5 rows:")
print(selected_columns.head())
```

```python
#Access the first 5 rows of the DataFrame.
first_five = df.head()

print("First 5 rows of the DataFrame:\n")
print(first_five)

print("\nShape of this subset:", first_five.shape)
```

```python
#Retrieve the row with CustomerID = 105.
row_105 = df[df["CustomerID"] == 105]

print("Row with CustomerID = 105:\n")
print(row_105)

print("\nType of result:", type(row_105))
print("Shape:", row_105.shape)
```

```python
#Get all customers where Age > 30.
customers_over_30 = df[df["Age"] > 30]

print("Customers where Age > 30:\n")
print(customers_over_30.head())

print("\nTotal customers with Age > 30:", customers_over_30.shape[0])
```

```python
#Find all purchases where PurchaseAmount > 250
high_purchases = df[df["PurchaseAmount"] > 250]

print("Purchases where PurchaseAmount > 250:\n")
print(high_purchases.head())

print("\nTotal purchases above 250:", high_purchases.shape[0])
```

```python
#Select all rows where Gender is 'F' and City is 'Los Angeles'.

filtered_data = df[(df["Gender"] == "F") & (df["City"] == "Los Angeles")]

print("Rows where Gender is 'F' and City is 'Los Angeles':\n")
print(filtered_data.head())

print("\nTotal matching rows:", filtered_data.shape[0])
```

```python
#Get all customers who have Gold membership.
gold_customers = df[df["Membership"] == "Gold"]

print("Customers with Gold membership:\n")
print(gold_customers.head())

print("\nTotal Gold members:", gold_customers.shape[0])

```

```python
#Find customers whose PurchaseAmount is missing (NaN).

missing_purchase = df[df["PurchaseAmount"].isna()]

print("Customers with missing PurchaseAmount:\n")
print(missing_purchase.head())

print("\nTotal customers with missing PurchaseAmount:", missing_purchase.shape[0])
```

```python
#Show all customers whose City is missing.
missing_city = df[df["City"].isna()]

print("Customers with missing City:\n")
print(missing_city.head())

print("\nTotal customers with missing City:", missing_city.shape[0])
```

```python
#Get all customers who are Male AND Age > 30.

male_over_30 = df[(df["Gender"] == "M") & (df["Age"] > 30)]

print("Male customers with Age > 30:\n")
print(male_over_30.head())

print("\nTotal Male customers with Age > 30:", male_over_30.shape[0])
```

```python
#Find customers who are Female OR have a PurchaseAmount > 300.

filtered_customers = df[(df["Gender"] == "F") | (df["PurchaseAmount"] > 300)]

print("Customers who are Female OR PurchaseAmount > 300:\n")
print(filtered_customers.head())

print("\nTotal matching customers:", filtered_customers.shape[0])
```

```python
#Select rows where Age is between 30 and 40 (inclusive).

age_30_40 = df[df["Age"].between(30, 40, inclusive="both")]

print("Customers with Age between 30 and 40 (inclusive):\n")
print(age_30_40.head())

print("\nTotal customers aged 30–40:", age_30_40.shape[0])
```

```python
#Get all customers whose Membership is NOT Bronze.
not_bronze = df[df["Membership"] != "Bronze"]

print("Customers whose Membership is NOT Bronze:\n")
print(not_bronze.head())

print("\nTotal customers not Bronze:", not_bronze.shape[0])
```

```python
#Select all rows where PurchaseAmount > 200 AND Membership = 'Gold'.
gold_high_spenders = df[(df["PurchaseAmount"] > 200) & (df["Membership"] == "Gold")]

print("Customers with PurchaseAmount > 200 AND Membership = 'Gold':\n")
print(gold_high_spenders.head())

print("\nTotal matching customers:", gold_high_spenders.shape[0])
```

```python
#Find customers in New York OR Chicago with Age < 35.
filtered_customers = df[
    (df["City"].isin(["New York", "Chicago"])) & 
    (df["Age"] < 35)
]

print("Customers in New York OR Chicago with Age < 35:\n")
print(filtered_customers.head())

print("\nTotal matching customers:", filtered_customers.shape[0])
```

```python
#Get all customers who do not have Membership info (NaN).
no_membership = df[df["Membership"].isna()]

print("Customers with missing Membership:\n")
print(no_membership.head())

print("\nTotal customers with missing Membership:", no_membership.shape[0])
```

```python
#Select top 3 customers with the highest PurchaseAmount.

top_3 = df.sort_values(by="PurchaseAmount", ascending=False).head(3)

print("Top 3 customers by PurchaseAmount:\n")
print(top_3)

print("\nShape:", top_3.shape)
```
