## Line Plot

```python
x=["Jan","Feb","Mar","Apr","May"]
y=[400,500,600,450,900]
plt.plot(x,y,marker="o", linestyle='-',color="g",label='Line 1')
plt.title("Line plot example")
plt.xlabel("months")
plt.ylabel("sales")
plt.legend()
plt.grid(True)
plt.show()
```

## Area Plot

```python
x=["Jan","Feb","Mar","Apr","May"]
y1=[400,500,600,450,900]
y2=[100,100,100,100,100]
plt.fill_between(x,y1,y2,color="skyblue",alpha=1.0,label='Area plot')
plt.title("Area  plot example")
plt.xlabel("months")
plt.ylabel("sales")
plt.legend()
plt.show()
```

## Bar Plot

```python
x=["Laptop","Mobile","Earphones","Tablets","Headphones"]
y=[500,300,600,250,100]
plt.bar(x,y,color='b',alpha=0.6)
plt.title("Bar Plot Example")
plt.xlabel("Electronics Gadgets")
plt.ylabel("Sales")
plt.grid(axis='y')
plt.show()
```

## Scatter Plot

```python
x=[1,2,3,4,5]
y=[10,14,18,16,12]
 
plt.scatter(x,y,marker="o",color="r",label='Data Points')
plt.title("Scatter  plot example")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.grid(True)
plt.show()
```

## Pie Plot

```python
x=["Laptop","Mobile","Earphones","Tablets","Headphones"]
y=[20,15,25,30,10] # sales percentage 
plt.pie(y,labels=x,autopct='%1.1f%%',startangle=90,colors=['gold','lightgreen','skyblue'])
#autopct--->it displays the percentage of each category on the pie chart with one decimal place
plt.title("Pie Chart example")
plt.axis('equal') # equal aspects rtio ensures that pie drawn as a circle
plt.show()#
```

```python
x=["Laptop","Mobile","Earphones","Tablets","Headphones"]
y=[20,15,25,30,10] # sales percentage 

explode=(0,0.2,0,0,0) # pullout mobile slice from pie chart
#0-->slice stays in place
# >0 or greater than 0-->slice moves out ward

plt.pie(y,labels=x,autopct='%1.1f%%',startangle=90,colors=['gold','lightgreen','skyblue'],explode=explode,shadow=True)

plt.title("Pie Chart example",fontsize=14,fontweight='bold')
plt.axis('equal') 
plt.show()
```
