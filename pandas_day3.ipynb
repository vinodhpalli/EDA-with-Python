# Data frame iteration methods

## iterrows()

```python
data={
    "ID":[1,2,3],
    "Name":["A","B","C"],
    "Marks":[85,90,75]
}
df=pd.DataFrame(data)
for index,row in df.iterrows():
    print(f"Index:{index},ID:{row["ID"]},Name:{row["Name"]},Marks:{row["Marks"]}")
```

## itertuples()

```python
data={
    "ID":[1,2,3],
    "Name":["A","B","C"],
    "Marks":[85,90,75]
}
df=pd.DataFrame(data)
for row in df.itertuples():
    print(f"Index:{row.Index}, ID:{row.ID Name:{row.Name},Marks:{row.Marks}")
```
