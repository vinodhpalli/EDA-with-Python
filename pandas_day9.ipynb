 ## multidimenstional aggregation using pivot table()

```python
import pandas as pd
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
data={
    "Region":["East","East","East","East","West","West","West","West","South","South","South","South"],
    "Product":["Laptop","Laptop","Mobile","Mobile","Laptop","Laptop","Tablet","Tablet","Mobile","Mobile","Tablet","Tablet"],
    "Sales":[800,500,600,400,1000,700,500,300,650,550,450,350],
    "Quantity":[5,3,4,2,6,4,3,2,5,3,4,2]
}
df=pd.DataFrame(data)
df
```

```python
pivot_index=pd.pivot_table(df,index="Region",values="Sales",aggfunc="sum")
pivot_index
```

```python
pivot_column=pd.pivot_table(df,index="Region",columns="Product",values="Sales",aggfunc="sum")
pivot_column
```

```python
pivot_fill=pd.pivot_table(df,index="Region",columns="Product",values="Sales",aggfunc="sum",fill_value=0)
pivot_fill
```

```python
pivot_fill=pd.pivot_table(df,index="Region",columns="Product",values="Sales",aggfunc=["sum","max"],fill_value=0)
pivot_fill
```

```python
pivot_count=pd.pivot_table(df,index="Region",columns="Product",values="Sales",aggfunc="count",fill_value=0)
pivot_count
```

```python
pivot_sum=pd.pivot_table(df,index="Region",columns="Product",values="Sales",aggfunc="sum",fill_value=0,margins=True)
pivot_sum
```

## categorical analysis using crosstab/crosstab

```python
data={
    "Employee":["John","James","David","Marry","Tom","Charlie"],
    "Department":["Sales","HR","Sales","IT","HR","IT"],
    "Gender":["Male","Female","Male","Female","Male","Male"],
    "Salary":[50000,60000,55000,70000,62000,68000]
}
df=pd.DataFrame(data)
df
```

```python
#count of employes per department
ct=pd.crosstab(index=df["Department"],columns=df["Gender"])
ct
```

```python
ct_sum=pd.crosstab(index=df["Department"],columns=df["Gender"],values=df["Salary"],aggfunc="sum")
ct_sum
```

```python
ct_multi=pd.crosstab(index=df["Department"],columns=df["Gender"],values=df["Salary"],aggfunc=["sum","mean"])
ct_multi
```

```python
ct_margins=pd.crosstab(index=df["Department"],columns=df["Gender"],
                     values=df["Salary"],aggfunc=["sum","mean"],margins=True)
ct_margins
```

```python
ct_norm=pd.crosstab(index=df["Department"],columns=df["Gender"],normalize="index")
ct_norm
```

## merging in pandas

```python
employees=pd.DataFrame({
    "EmpID":[1,2,3,4],
    "EmpName":["A","B","C","D"],
    "DeptID":[101,102,103,104]
})
departments=pd.DataFrame({
    "DeptID":[101,102,103,105],
    "DeptName":["HR","IT","Finane","Marketing"]
})
print(employees)
print("--"*20)
print(departments)
```

### inner join 

```python
#returns common row only
inner_join=pd.merge(employees,departments,on="DeptID",how="inner")
print(inner_join)
```

## left join

```python
#returns common row only
left_join=pd.merge(employees,departments,on="DeptID",how="left")
print(left_join)
```

## right join

```python
#all departments+matching emps
right_join=pd.merge(employees,departments,on="DeptID",how="right")
print(right_join)
```

## outer join

```python
#all records from both df's
outer_join=pd.merge(employees,departments,on="DeptID",how="outer")
print(outer_join)
```
