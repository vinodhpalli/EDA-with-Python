# datatype conversion

```python
 import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.DataFrame({
    "Age":["23","25","21"],
    "Salary":[1000.5,2000.0,1500.75],
    "Employed":[1,0,1]
})
df.info()
```

```python
df.info()
```

```python
#convert age from object to int
df["Age"]=df["Age"].astype(int)
df.dtypes
```

```python
df["Salary"]=df["Salary"].astype(int)
df.dtypes
```

```python
df["Employed"]=df["Employed"].astype(int)
df.dtypes
```

## pd.to_numeric

```python
df["Salary"]=df["Salary"].astype(str)
print(df.dtypes)
print("---"*15)
df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
print(df.dtypes)
```

## pd.to_datetime

```python
df["JoinDate"]=pd.to_datetime(["2025-09-09","2024-01-01","2023-12-15"])
df.dtypes
```

## pd.to_timedelta()

```python
df["Experience"]=pd.to_timedelta(["365 days","730 days","1095 days"])
df.dtypes
```

## pd.series()

```python
df["Department"]=pd.Series(["HR","IT","Finance"]).astype("category")
df.dtypes
```
