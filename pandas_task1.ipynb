# üêº Pandas DataFrame ‚Äì Practice Questions


```python
#Create a Pandas DataFrame from a list of lists representing student ID, name, and score.
data={
    'ID':[0,1,2,3],
    'name':["A","B","C","D"],
    'score':[85,92,78,90]
}
df=pd.DataFrame(data)
df
```

```python
#Create a Pandas DataFrame from a dictionary of lists representing product, price, and quantity.
data={
    "product":["Soap","Paste","Brush"],
    "price":[10,20,25],
    "Quantity":[1,3,4]
}
df=pd.DataFrame(data)
df
```

```python
#Create a Pandas DataFrame from a NumPy array representing employee data.
data={
    "Employee_ID":[101,102,103],
    "Employee_name":["vinodh","viswes","sarojini"],
    "salary":[10000,20000,30000]
}
df=pd.DataFrame(data)
df
```

```python
#Assign appropriate column names to the DataFrame created from the list of lists.
data = [
    [1, "Alice", 85],
    [2, "Bob", 78],
    [3, "Charlie", 92]
]

df = pd.DataFrame(data, columns=["ID", "Name", "Marks"])

print(df)
```

```python
#Assign column names to the DataFrame created from the NumPy array.
arr = np.array([
    [1, "Alice", 85],
    [2, "Bob", 78],
    [3, "Charlie", 92]
])
df = pd.DataFrame(arr, columns=["ID", "Name", "Marks"])
print(df)
```

```python
#Print the shape of each DataFrame created.
arr = np.array([
    [1, "Alice", 85],
    [2, "Bob", 78],
    [3, "Charlie", 92]
])
print(arr.shape)
```

```python
#Print the column names of each DataFrame.
df =pd.DataFrame([
    [1, "Alice", 85],
    [2, "Bob", 78],
    [3, "Charlie", 92]
])
print(df.columns)
```

```python
#Print the index values of one DataFrame.
data = [
    [1, "Alice", 85],
    [2, "Bob", 78],
    [3, "Charlie", 92]
]

df = pd.DataFrame(data, columns=["ID", "Name", "Marks"])
print(df.index)
```

```python
#Display the complete structure (rows and columns) of a DataFrame using attributes.
print(df.shape)      # (rows, columns)
print(df.index)      # row labels
print(df.columns)    # column labels
print(df.dtypes)     # data type of each column
```
