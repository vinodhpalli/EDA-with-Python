## Apply()

```python
data={
    "Name":["Aman","Bob","Charlie","David"],
    "Department":["HR","IT","Finance","IT"],
    "Salary":[50000,60000,55000,70000],
    "Joining_Year":[2019,2020,2018,2021]
}
df=pd.DataFrame(data)
df
```

```python
Salaries=[1000,2000,3000]
updated=list(map(lambda x:x+500,Salaries))
print(updated)
```

```python
df["Updated_salaries"]=df["Salary"].apply(lambda x:x+5000)
df
```

```python
def salary_category(Salary):
    if Salary>60000:
        return "High"
    else:
        return "low"
df["Salary_category"]=df["Salary"].apply(salary_category)
df
```

```python
import datetime

current_year=datetime.datetime.now().year
df["Experience"]=df.apply(lambda row : current_year - row["Joining_Year"],axis=1)
df
```

```python
df["Name_upper"]=df["Name"].apply(str.upper)
df
```

```python
df["Dept_len"]=df["Department"].apply(len)
df
```

```python
def added_salary(Salary):
    Salary+=60000
    return Salary
df["Updated_salaries"]=df["Salary"].apply(added_salary)
df
```

## applymap()

```python
data={
    "Name":["Aman","Bob","Charlie","David"],
    "Department":["HR","IT","Finance","IT"],
    "Salary":[50000,60000,55000,70000],
    "Bouns":[5000,6000,5500,6000]
}
df=pd.DataFrame(data)
df
```

```python
def add_numeric(x):
    if isinstance(x,(int,float)):
        return x + 1000
    return x
df2=df.applymap(add_numeric)
df2
```

```python
def str_update(x):
    if isinstance(x,str):
        return x.upper()
    return x
df3=df.applymap(str_update)
df3
    
```

```python
def numeric(x):
    if isinstance(x,(int,float)):
        if x>60000:
            return "High"
    return x
df4=df.applymap(numeric)
df4
```

### Rolling() And Expanding()

```python
sales=pd.Series([10,20,30,40,50])
sales
```

```python
print("Rolling:\n",sales.rolling(3).mean())
```

```python
print("Expanding:\n",sales.expanding().mean())
```
