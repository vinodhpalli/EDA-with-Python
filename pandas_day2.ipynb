```python
import pandas as pd
import numpy as np
```

```python
# create a series from dictonary
data={"A":85,"B":92,"C":78,"D":90,"E":88}

marks=pd.Series(data)
print(marks)
```

```python
#sort a series
data={"A":85,"D":90,"C":78,"B":92,"E":88}
marks=pd.Series(data)
print(marks)
print("--"*20)
#sort by index
print(marks.sort_index())
print("--"*20)
#sort by values
print(marks.sort_values()) # ascending order
print("--"*20)
print(marks.sort_values(ascending=False))#decending order
```

```python
#reordering the index in series
data={"A":85,"D":90,"C":78,"B":92,"E":88}
marks=pd.Series(data)
print(marks)
print("--"*20)

marks=pd.Series(data,index=["A","B","C","D","E"])
print(marks)
```

## Loc and Iloc

```python
marks=pd.Series(["A","B","C","D"],index=[101,102,103,104])
print(marks)
print("--"*20)
print('loc values') # manually generate index values
print(marks.loc[101])
print(marks.loc[102])
print(marks.loc[103])
print(marks.loc[104])
#print(marks.loc[1])#---->it shows error
print("--"*20)
print('iloc values') # original index values
print(marks.iloc[0])
print(marks.iloc[1])
print(marks.iloc[2])
print(marks.iloc[3])
print("--"*20)
print(marks.loc[102:104])
print("--"*20)
print(marks.iloc[1:4])
```

```python
## loc iloc on dataframe
data={
    'name':["A","B","C","D","E"],
    'age':[23,25,22,24,26],
    'dept':["HR","IT","FINANCE","IT","HR"],
    'Salry':[30000,35000,28000,40000,45000]
}
df=pd.DataFrame(data,index=["E1","E2","E3","E4","E5"])
df
print(df.loc["E1"])
print("--"*20)
print(df.loc["E2"])
print("--"*20)
print(df.loc["E3"])
print("--"*20)
print(df.loc["E4"])
print("--"*20)
print(df.loc[["E1","E3"]])
print("--"*20)
print(df.loc["E2":"E4"])
print("--"*20)
print(df.loc[["E2","E4"],["name","Salry"]])
print("--"*20)
print(df.loc[df['Salry']>35000])
print("--"*20)

#iloc
print("iloc examples:")
print(df.iloc[0])
print("--"*20)
print(df.iloc[1])
print("--"*20)
print(df.iloc[[0,3]])
print("--"*20)
print(df.iloc[0:4])
print("--"*20)
print(df.iloc[[0,2],[0,3]])
print("--"*20)
high_salary=df['Salry']>35000  #----->0,2-->e1,e2 rows \\ 0,3--->name ,salary(colum)
print(df.iloc[high_salary.values])
```

## DataFrame operations:column& row manipulation

```python
import pandas as pd
import numpy as np
```

```python
data_a={
    'name':["A","B","C","D","E"],
    'age':[23,25,22,24,26],
    'marks':[85,92,78,90,88],
    'city':["delhi","mumbai","chennai","bengalor","hyderabad"]
}
df=pd.DataFrame(data_a)
df
```

```python
# column selection
print(df['name'])
print("--"*20)
print(df[['name','age']])
```

```python
# add new column
df["grade"]=["B","A","C","A+","B+"]
df
```

```python
# sorting
print(df.sort_values(by="age")) #ascending
print("--"*20)
print(df.sort_values(by="age",ascending=False)) #decending
```

```python
# column deletion
df=df.drop('age',axis=1)
df
```

```python
#row selection 
print(df.iloc[2])
print("---"*10)
print(df.iloc[2])   
    
```

```python
# add new row
new_row={
    "name":'F',"age":23,"marks":92,"grade":"A"
}
df.loc[5]=new_row
print(df)
```

```python
# row deletion
df=df.drop(5,axis=0)
print(df)
```

```python
#give  all statitical values numerical values
print(df.describe())
print("--"*20)
print(df.describe(include='all')) #give all statitical values for all coumns
```

## groupby()

```python
data={
    'name':["A","B","C","D","E"],
    'age':[25,25,24,24,25],
    'marks':[85,92,78,90,88],
    'city':["delhi","mumbai","chennai","bengalor","hyderabad"]
}
df=pd.DataFrame(data)
df
```
