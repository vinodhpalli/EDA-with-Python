```python
word_to_num = {
    "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
    "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
    "ten": 10, "eleven": 11, "twelve": 12,
    "thirteen": 13, "fourteen": 14, "fifteen": 15,
    "sixteen": 16, "seventeen": 17, "eighteen": 18,
    "nineteen": 19, "twenty": 20, "twenty-one": 21,
    "twenty-two": 22, "twenty-three": 23
}

df["hour_of_day"] = (
    df["hour_of_day"]
    .astype(str)
    .str.strip()
    .str.lower()
)
df["hour_of_day"] = df["hour_of_day"].replace(word_to_num)
df["hour_of_day"] = pd.to_numeric(df["hour_of_day"], errors="coerce").astype("Int64")
print("Column data type after conversion:")
print(df["hour_of_day"].dtype)
print("\nSample values:")
print(df["hour_of_day"].head(10))
```

```python
#Convert the money column into float values (fix entries like "thirty eight").
word_to_num = {
    "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
    "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
    "ten": 10, "eleven": 11, "twelve": 12, "thirteen": 13,
    "fourteen": 14, "fifteen": 15, "sixteen": 16,
    "seventeen": 17, "eighteen": 18, "nineteen": 19,
    "twenty": 20, "thirty": 30, "forty": 40, "fifty": 50,
    "sixty": 60, "seventy": 70, "eighty": 80, "ninety": 90
}
def words_to_number(text):
    if pd.isna(text):
        return None
    text = str(text).strip().lower()
    try:
        return float(text)
    except:
        pass
    
    parts = text.replace("-", " ").split()
    total = 0
    for part in parts:
        if part in word_to_num:
            total += word_to_num[part]
        else:
            return None 
    return float(total)
df["money"] = df["money"].apply(words_to_number)

print("Column data type after conversion:")
print(df["money"].dtype)
print("\nSample values:")
print(df["money"].head(10))
```

```python
df_raw=pd.read_csv("Coffee_sales_uncleaned.csv")
df

print("===== BEFORE CLEANING =====")
print(df_raw.describe())
print("\n")

# --- Cleaning numeric columns ---

# Convert hour_of_day
hour_map = {
    "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
    "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
    "ten": 10, "eleven": 11, "twelve": 12,
    "thirteen": 13, "fourteen": 14, "fifteen": 15,
    "sixteen": 16, "seventeen": 17, "eighteen": 18,
    "nineteen": 19, "twenty": 20, "twenty-one": 21,
    "twenty-two": 22, "twenty-three": 23
}

df_clean = df_raw.copy()

df_clean["hour_of_day"] = (
    df_clean["hour_of_day"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace(hour_map)
)

df_clean["hour_of_day"] = pd.to_numeric(df_clean["hour_of_day"], errors="coerce")

money_map = {
    "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
    "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
    "ten": 10, "eleven": 11, "twelve": 12, "thirteen": 13,
    "fourteen": 14, "fifteen": 15, "sixteen": 16,
    "seventeen": 17, "eighteen": 18, "nineteen": 19,
    "twenty": 20, "thirty": 30, "forty": 40, "fifty": 50,
    "sixty": 60, "seventy": 70, "eighty": 80, "ninety": 90
}

def convert_money(x):
    if pd.isna(x):
        return None
    x = str(x).strip().lower()
    try:
        return float(x)
    except:
        parts = x.replace("-", " ").split()
        total = 0
        for p in parts:
            if p in money_map:
                total += money_map[p]
            else:
                return None
        return float(total)

df_clean["money"] = df_clean["money"].apply(convert_money)

print("===== AFTER CLEANING =====")
print(df_clean.describe())
```
