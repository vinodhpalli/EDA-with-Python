## selecting and swaping columns()

```python
df=pd.DataFrame({
    "Name":["A","B","C"],
    "Age":[23,25,21],
    "City":["Delhi","Mumbai","Chennai"]
})
df
```

```python
print(df["Name"])
print("--"*20)
print(df[["Name","City"]])
print("--"*20)
print(df.iloc[:,1])#first column
print("--"*20)
print(df.iloc[:,[0,2]])#first column and third column
print("--"*20)
print(df.iloc[:,0:2])#first two columns 


```

## swaping-->rearrange the order

```python
df1=df[["Name","City","Age"]]
print(df1)
print("--"*20)
#swap two columns(2nd method)
cols=list(df.columns)
cols[1],cols[2]=cols[2],cols[1]
df2=df[cols]
print(df2)
#3rd method-->reindex
df3=df.reindex(columns=["City","Name","Age"])
print(df3)
df4=df.iloc[:,[2,0,1]]
print(df4)
```
