## Histogram 

```python
data=[10,15,15,20,20,25,25,30]
plt.hist(data,bins=5,color='green',alpha=0.7)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()

```

```python
df=pd.read_csv("Data.csv")
df
```

## Line Plot

```python
plt.figure(figsize=(5,3)) # width,height
plt.plot(df["Name"],df["Age"],marker="o", linestyle='-',color="g")
plt.title("Line plot example")
plt.xlabel("Name")
plt.ylabel("Age")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Area Plot

```python
plt.figure(figsize=(5,3))
plt.fill_between(df["Name"],df["Salary"],color="skyblue",alpha=0.8)
plt.plot(df["Name"],df["Salary"],marker="o",color="g")
plt.title("Salary of Employees")
plt.xlabel(" Employee Name")
plt.ylabel("Salary")
plt.show()
```

## Bar Plot

```python
plt.figure(figsize=(6,5))
plt.bar(df["Name"],df["Salary"],color='g',alpha=0.6)
plt.title("Bar Plot Example")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(6,3))
plt.scatter(df["Salary"],df["Rating"],marker="o",color="r",label='Data Points')
plt.title("Salary vs Rating")
plt.xlabel("Salary")
plt.ylabel("Rating")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.hist(df["Age"],bins=3,color='purple',alpha=0.7)
plt.title("Histogram-Age Distrubution Example")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

```python


plt.pie(df["Salary"],labels=df["Name"],autopct='%1.1f%%',startangle=90)

plt.title("Pie Chart example")
 
plt.show()
```
