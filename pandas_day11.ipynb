## date time manipulation

```python
data={
    "Student":["Ali","Bob","Charlie","David"],
    "JoinDate":["2023-01-15","2023-03-22","2023-05-10","2023-07-01"],
    "Lastlogin":["2023-09-01 10:30:00","2023-09-05 15:45:00","2023-09-07 09:15:00","2023-09-10 20:00:00"]
}
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df["JoinDate"]=pd.to_datetime(df["JoinDate"])
df["Lastlogin"]=pd.to_datetime(df["Lastlogin"])
df.info()
```

### extract year month day

```python
print(df["JoinDate"].dt.year) # extraxt year from date
print("----"*10)
print(df["JoinDate"].dt.month) # extraxt month from date
print("----"*10)
print(df["JoinDate"].dt.day) # extraxt day from date
print("----"*10)
print(df["JoinDate"].dt.weekday) # extraxt year from dat
print("----"*10)
print(df["JoinDate"].dt.day_name()) # extraxt day name from date
print("----"*10)
print(df["JoinDate"].dt.month_name()) # extraxt monthname from dat
print("----"*10)
```

### difference b/w dates

```python
df["Lastlogin"]-df["JoinDate"]
```

```python
(df["Lastlogin"]-df["JoinDate"]).dt.days
```

```python
print(df["Lastlogin"]+pd.Timedelta(days=30)) # add extra 30 days
print("----"*10)
print(df["Lastlogin"]-pd.Timedelta(days=30)) # substract extra 30 days
print("----"*10)
```

```python
print(df["Lastlogin"].dt.hour)
print("----"*10)
print(df["Lastlogin"].dt.minute)
print("----"*10)
print(df["Lastlogin"].dt.second)
print("----"*10)
```
