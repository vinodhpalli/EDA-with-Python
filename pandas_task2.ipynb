 ## Dataset Creation 


```python
import pandas as pd
```

```python
data = {
    "emp_id": [101, 102, 103, 104, 105],
    "name": ["Asha", "Ravi", "Kiran", "Neha", "Arjun"],
    "department": ["IT", "HR", "IT", "Finance", "HR"],
    "salary": [60000, 45000, 70000, 50000, 48000],
    "experience": [3, 2, 5, 4, 1]
}

df = pd.DataFrame(data)
df
```

### .loc vs .iloc


```python
#Task 1.1
#Using .loc, select employees who belong to the IT department and display only name and salary.
df.loc[df["department"] == "IT", ["name", "salary"]]
```

```python
#Task 1.2
#Using .iloc, fetch the first 3 rows and first 2 columns.
df.iloc[:3, :2]
```

##  Renaming Columns


```python
#Task 2.1
#Rename:
#emp_id ‚Üí employee_id
#name ‚Üí employee_name
df1=df.rename(columns={"emp_id":"employee_id","name":"employee_name"})
df1
```

## isin() and not in


```python
import pandas as pd
```

```python
data = {
    "emp_id": [101, 102, 103, 104, 105],
    "name": ["Asha", "Ravi", "Kiran", "Neha", "Arjun"],
    "department": ["IT", "HR", "IT", "Finance", "HR"],
    "salary": [60000, 45000, 70000, 50000, 48000],
    "experience": [3, 2, 5, 4, 1]
}

df = pd.DataFrame(data)
df
```

```python
#### Task 6.1
#Select employees whose department is in:
#["IT", "Finance"]
df[df["department"].isin(["IT", "Finance"])]
```

```python
#Task 6.2
#Select employees whose names are NOT in:
#["Asha", "Arjun"]

df[~df["name"].isin(["Asha", "Arjun"])] 
```
