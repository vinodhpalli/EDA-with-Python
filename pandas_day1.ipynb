### 1.series

```python
data=[10,20,30,40,50]
s=pd.Series(data)
print(s)
```

```python
s1=pd.Series([10,20,30],index=['a','b','c'])
print(s1)
```

```python
marks=pd.Series([67,56,89,75,98],
                index=['john','prem','vinodh','vis','sar'])
print(marks)
```

```python
#acess by position
print(marks['vinodh'])
```

```python
# use slicing
print(marks[0:3])
```

```python
#vectorized operations
print(marks+5)
```

## aggregate functions()

```python
## aggregate functions-->sum(),min(),max(),mean(),count()
print(marks.sum())
print(marks.mean())
print(marks.min())
print(marks.max())
print(marks.count())
```

```python
# membership operator
print("manohar"in marks)
print("vinodh"in marks)
print("manohat"not in marks)
```

```python
# filtering
print(marks[marks>80])
```

```python
marks['sar']=90
print(marks)
```

```python
# get index and values
print(marks.index)
print(marks.values)
```

```python
print(marks.dtype)
print(marks.shape)
print(len(marks))
print(marks.head(3))
print(marks.head())#top5 records
print(marks.tail())#bottom5 records
```
