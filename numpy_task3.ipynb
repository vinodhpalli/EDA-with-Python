# üîπ Part B: Boolean Indexing


```python
#5Ô∏è‚É£ Condition-Based Filtering
#Input
#A = np.array([5, 12, 17, 3, 8, 20])

#Expected Output
#Values > 10        ‚Üí [12, 17, 20]
#Values ‚â§ 8         ‚Üí [5, 3, 8]
A = np.array([5, 12, 17, 3, 8, 20])
con=A>10
print(A[con])
con1=A<=8
print(A[con1])
```

```python
#6Ô∏è‚É£ Multiple Boolean Conditions
#Input
#A = np.array([10, 15, 20, 25, 30, 35])

#Expected Output
#[20, 25, 30, 35]
A = np.array([10, 15, 20, 25, 30, 35])
con=A>15
print(A[con])
```

```python
#7Ô∏è‚É£ Boolean Filtering + Counting
#Input
#A = np.array([3, -2, 0, 5, -7, 8])

#Expected Output
#Non-negative values ‚Üí [3, 0, 5, 8]
#Negative count      ‚Üí 2
A = np.array([3, -2, 0, 5, -7, 8])
con=A>0
print("Non-negative values:",A[con])
negative_sum=np.sum(A < 0)
print("Non-negative values:",negative_sum)
```

```python
#8Ô∏è‚É£ Boolean Replacement
#Input
#A = np.array([1, 2, 3, 4, 5, 6, 7])

#Expected Output
#[-1, 2, -1, 4, -1, 6, -1]
arr= np.array([1, 2, 3, 4, 5, 6, 7])
arr[[0,2,4,6]]=[-1]
print(arr)
```

# üîπ Part D: Boolean Indexing vs np.where()


```python
#1Ô∏è‚É£3Ô∏è‚É£ Filtering vs Transforming
#Input

#Expected Output
#Boolean indexing ‚Üí [9, 7, 12]
#np.where         ‚Üí [0, 1, 0, 1, 1, 0]
A = np.array([4, 9, 2, 7, 12, 3])
odd_values = A[A % 2 != 0]
print("Boolean indexing ‚Üí", odd_values)
mask = np.where(A % 2 != 0, 1, 0)
print("np.where         ‚Üí", mask)
```

```python
#1Ô∏è‚É£4Ô∏è‚É£ Replace Negatives (Two Methods)
#Input
#A = np.array([-5, 10, -3, 8, -1])

#Expected Output
#[0, 10, 0, 8, 0]
A = np.array([-5, 10, -3, 8, -1])
mask = np.where(A>0,A,0)
print("np.wher‚Üí", mask)
```

```python
#1Ô∏è‚É£5Ô∏è‚É£ Labeling Data with np.where()
#nput
#scores = np.array([45, 60, 30, 85, 90])

#Expected Output
#Filtered scores ‚Üí [60, 85, 90]
#Labels          ‚Üí ["Fail", "Pass", "Fail", "Pass", "Pass"]
A= np.array([45, 60, 30, 85, 90])
con=A>50
print("Filtered scores:",A[con])
labels = np.where(A >= 50, "Pass", "Fail")
print("labels:",labels)
```

# üîπ Part E: Broadcasting


```python
#1Ô∏è‚É£6Ô∏è‚É£ Row-wise Broadcasting
#Input
#M = np.array([[1, 2, 3],
               #[4, 5, 6]])
#A = np.array([10, 20, 30])

#Addition     ‚Üí [[11, 22, 33],
                #[14, 25, 36]]

#Subtraction  ‚Üí [[-9, -18, -27],
               # [-6, -15, -24]]
M = np.array([[1, 2, 3],
              [4, 5, 6]])
A = np.array([10, 20, 30])
print("Addition:",M+A)
print("--"*10)
print("subtraction:",M-A)
```

```python
#1Ô∏è‚É£7Ô∏è‚É£ Row Normalization
#nput
#M = np.array([[2, 4],
 #             [6, 8]])

#Expected Output
#[[1, 2],
#[3, 4]]


M = np.array([[2, 4],
            [6, 8]])

row_normalized = M / M[:, [0]]
print(row_normalized)
```

```python
#1Ô∏è‚É£8Ô∏è‚É£ Scalar Broadcasting
#Input
#A = np.array([10, 20, 30])

#Expected Output
#[15, 25, 35]
A = np.array([10, 20, 30])
print(A+5)
```

```python
#1Ô∏è‚É£9Ô∏è‚É£ Boolean vs np.where() on 2D
#Input
#M = np.array([[5, 10, 15],
 #             [20, 25, 30]])

#Expected Output
#[[5, 10, 15],
 #[20, 20, 20]]

M = np.array([[5, 10, 15],
              [20, 25, 30]])

result_where = np.where(M > 20, 20, M)
print(result_where)

```

```python
#2Ô∏è‚É£0Ô∏è‚É£ Broadcasting Concept Check
#Input
#M = np.array([[1, 2],
              #[3, 4],
             # [#5, 6]])
#A = np.array([10, 20])

#Expected Output
#[[11, 22],
# [13, 24],
# [15, 26]]
M = np.array([[1, 2],
              [3, 4],
              [5, 6]])
A = np.array([10, 20])

result = M + A
print(result)
```
