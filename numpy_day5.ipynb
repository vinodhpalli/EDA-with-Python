# statistics

## mean()

```python
import numpy as np
```

```python
arr=np.array([50,70,30,90,60])
average=np.mean(arr)
print("mean value:",average)
```

```python
arr2d=np.array([[1,2,3],
                [4,5,6]])
print("mean of all  values:",np.mean(arr2d))
print("mean of each row:",np.mean(arr2d,axis=0))
print("mean of each column:",np.mean(arr2d,axis=1))
```

## median()

```python
arr=np.array([50,70,30,90,60])
med=np.median(arr)
print("median value:",med)
```

```python
arr2d=np.array([[1,2,3],
                [4,5,6]])
print("median of all  values:",np.median(arr2d))
print("median of each row:",np.median(arr2d,axis=0))
print("median of each column:",np.median(arr2d,axis=1))
```

## varience()

```python
marks=np.array([50,70,30,90,60])
print("varience:",np.var(marks))
```

```python
marks=np.array([[1,2,3],
                [4,5,6]])
print("varience:",np.var(marks))
print("varience of each row:",np.var(marks,axis=0))
print("varienceof each column:",np.var(marks,axis=1))
```

## standard devition

```python
marks=np.array([50,70,30,90,60])
print("standad devition:",np.std(marks))
```

```python
marks=np.array([[1,2,3],
                [4,5,6]])
print("standad devition:",np.std(marks))
print("standad devition of each row :",np.std(marks,axis=0))
print("standad devition of each column:",np.std(marks,axis=1))
```

## co-relation co-efficient

```python
#positive
x=[1,2,3,4,5]
y=[2,4,6,8,10]
corr=np.corrcoef(x,y)
print(corr)
print("--------------")
#negative correlation
x1=[1,2,3,4,5]
y1=[10,8,6,4,2]
corr1=np.corrcoef(x1,y1)
print(corr1)
```

# Array manipulation

## concatenation()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.concatenate((a,b))
print(c)
```

```python
a=np.array([[1,2,3],
         [4,5,6]])
b=np.array([[7,8,9],
            [10,11,12]])
c=np.concatenate((a,b))
print(c)
print("--------------")
#row wise
row_con=np.concatenate((a,b),axis=0)
print(row_con)
print("--------------")
#column wise
col_con=np.concatenate((a,b),axis=1)
print(col_con)
```

## stack()

### vstack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.vstack((a,b))
print(c)
```

```python
a=np.array([[1,2,3],
         [4,5,6]])
b=np.array([[7,8,9],
            [10,11,12]])
c=np.vstack((a,b))
print(c)
```

## Hstack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.hstack((a,b))
print(c)
```

```python
a=np.array([[1,2,3],
         [4,5,6]])
b=np.array([[7,8,9],
            [10,11,12]])
c=np.hstack((a,b))
print(c)
```

### column stack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.column_stack((a,b))
print(c)
```

```python
a=np.array([[1,2,3],
         [4,5,6]])
b=np.array([[7,8,9],
            [10,11,12]])
c=np.column_stack((a,b))
print(c)
```

## sorting()

```python
arr=np.array([5,1,4,3,2])
sorted_arr=np.sort(arr)
print("accending order:",sorted_arr)
print("decending order:",sorted_arr[::-1])
```

```python
import numpy as np
a=np.array([[3,1,4],
            [5,6,2]])
sorted_arr=np.sort(a)
print("accending order:",sorted_arr)
print("decending order:",sorted_arr[::-1])
```

## arg sort()

```python
marks=np.array([40,10,30,20])   # index position=[0,1,2,3]
sorted_marks=np.sort(marks)
sorted_indices=np.argsort(marks )#--->it gives index postions
print("original:",marks)
print("sorted values:",sorted_marks)
print("argsorted values:",sorted_indices)
```

```python
a=np.array([[3,1,4],
            [5,6,2]])
print("argsorted column values:\n",np.argsort((a,axis=0))
print("argsorted row values:\n",np.argsort((a,axis=1))
```
