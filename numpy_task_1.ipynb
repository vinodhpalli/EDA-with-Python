#Explain why NumPy arrays are stored in contiguous blocks of memory and why this is important.
NumPy arrays are stored in contiguous blocks of memory to enable efficient data handling and computation.

Storage Mechanism
NumPy allocates a single, continuous chunk of memory for an array's data buffer, where elements of a uniform data type (dtype) are placed sequentially without gaps or pointers between them. This contrasts with Python lists, which store references to separate objects scattered in memory. Arrays can be C-contiguous (row-major order, default) or F-contiguous (column-major), but both maintain adjacency for their respective traversal patterns.

Explain the difference between selecting elements by index and selecting elements by a condition (boolean indexing).

Indexing by index and boolean indexing in NumPy arrays select elements differently: the former uses specific positions, while the latter filters based on conditions.

Index Selection
Index selection accesses elements at exact integer positions, like arr[2] for the third item or arr[1:4] for a slice. It requires knowing positions upfront and returns a view or copy of those specific locations. This works for single values, slices, or lists of indices (fancy indexing).

Boolean Indexing
Boolean indexing uses a matching boolean array (or condition) where True selects the element and False skips it, e.g., arr[arr > 5] picks all values greater than 5. NumPy broadcasts the mask across dimensions if needed, flattening results into a new array of selected elements. It's vectorized for conditions like arr[(arr > 3) & (arr < 7)

```python
#From an array [3, 7, 2, 9, 5, 10], select only the elements that are greater than 5.
arr=np.array([3, 7, 2, 9, 5, 10])
v=arr[arr>5]
print(v)
```

Explain whether creating a sub-array using slicing creates a view of the original array or a copy.
NumPy slicing typically creates a view of the original array, not a copy, to save memory and time.

View Behavior
Basic slicing (e.g., arr[1:3], arr[::2], or arr[:, 1]) produces a view that references the same underlying data buffer. Changes to the sliced array affect the original, like sliced[0] = 99 updating arr. This works because slices maintain contiguous memory strides when possible.

What is the most suitable data type for storing RGB pixel values (0–255)?

uint8 (8-bit unsigned integer) is the most suitable data type for storing RGB pixel values in the range 0–255.

Why uint8?
Each RGB channel fits perfectly in uint8, which stores integers from 0 to 255 using just 1 byte per channel (3 bytes per pixel). This is the standard for image formats (JPEG, PNG) and libraries like PIL, OpenCV, and matplotlib expect this dtype for pixel data.

Memory Efficiency
A 1920×1080 RGB image uses 6.2 MB with uint8 vs. 12.4 MB with float32 or 24.8 MB with float64. No precision loss since pixels are inherently 8-bit integers.

# X. Exploring Documentation


### Explore the NumPy documentation and learn about all the array methods and functions that we have studied in class so far. Write a short note on what each method does and how it can be used.

Specifying dtype during array creation and various indexing methods are key NumPy concepts from class discussions and documentation.

Array Creation with dtype
dtype sets element type (e.g., np.uint8, np.float32), controlling memory, precision, and operations. Use via np.array(data, dtype=np.int32) to optimize storage (e.g., RGB pixels as uint8) and prevent promotion (e.g., int division stays int).

Basic Indexing
Accesses single elements (arr[2]) or subarrays by integer positions (0-based, negative from end). Returns scalar or view; e.g., arr[1:4] selects indices 1-3.

Slicing
Extracts contiguous ranges with arr[start:stop:step] (defaults: start=0, stop=len, step=1). Always creates views for efficiency; e.g., arr[::2] every other element. Negative step reverses.

Boolean Indexing
Filters via condition mask (arr[arr > 5]), returning flattened copy of matching elements. Vectorized; e.g., arr[(arr > 3) & (arr < 7)]. Shape matches True count.

Fancy Indexing
Uses integer arrays (arr[[0,2,4]]) or multi-dim indices for arbitrary selection, always creating copies. Broadcasting applies; combines with slices.

Views vs Copies
Basic slicing yields views (shared memory, changes propagate); fancy/boolean yield copies. Check with arr.base; use .copy() for explicit copy
